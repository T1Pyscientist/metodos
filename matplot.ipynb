```python
def graficar_ecuaciones(b_1:int, b_2:int) -> None:
    x = np.linspace(-1, 5, 7)
    y_1 = (3 * x - b_1) / 2
    y_2 = (6 * x - b_2) / 4

    fig, ax = plt.subplots()
    ax.plot(x, y_1)
    ax.plot(x, y_2)
    plt.show()



graficar_ecuaciones(1, 2)
```

```python
import math
```

```python
def scale_a_b(v:Vector, alpha:float, beta:float) -> Vector:
    return Vector([v.componentes[0] * alpha, v.componentes[1] * beta])

def rotate_v(v:Vector, alpha:float)-> Vector:
    x = v.componentes[0]
    y = v.componentes[1]

    new_x = round(math.cos(alpha)*x - math.sin(alpha)*y)
    new_y = round(math.sin(alpha)*x + math.cos(alpha)*y)

    return Vector([new_x, new_y])
```

```python
rotate_v(Vector([0, 1]), math.pi/2)
```

```python
ax = plt.subplot()

x1, y1 = scale_a_b(Vector([1, 1]), 2, 1).componentes
x2, y2 = rotate_v(Vector([0, 1]), math.pi/2).componentes
ax.scatter([x1, x2], [y1, y2])

plt.show()
```

```python
def matrixR(v:Vector, k:int) -> Matrix:
    n = v.dimension
    valores = [[1  if j == i else 0 for j in range(n)] for i in range(n)]
    valores[k][k] = 0
    return Matrix(valores)

matrixR(Vector([1, 2, 3]), 1)
```
